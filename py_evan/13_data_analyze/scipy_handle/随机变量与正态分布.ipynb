```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import scipy.stats as st

# 2.2.1 创建随机变量：设置size、设置loc期望值、设置scale方差
s1 = st.norm.rvs(loc=0, scale=1, size=1000)

s2 = st.norm.rvs(loc=0, scale=2, size=1000)

s3 = st.norm.rvs(loc=0, scale=3, size=1000)

for s in [s1, s2, s2]:
    sns.distplot(s, bins=1000)
plt.legend(labels=[1, 2, 3])
```
