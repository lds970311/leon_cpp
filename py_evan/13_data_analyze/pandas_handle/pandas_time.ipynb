## 3.3.2 Resample函数

- 计数、均值、方差、累加、累乘
- 周期转换
- 数据验证：for循环 vs内置函数

```python
import pandas as pd
import pandas_datareader as pdr

frame = pdr.get_data_fred('GS10')
type(frame)  # pandas.core.frame.DataFrame
# 加载新数据
frame2 = pdr.get_data_fred('GS5')
frame['GS5'] = frame2  # 值和index的精确匹配
frame['mean'] = (frame['GS10'] + frame['GS5']) / 2
frame.resample(rule='Y').last()  # 查看每年最后1个月的数据
```

```python
# 计算：均值、方差、最大值、最小值、计数等常见的描述性指标
frame.resample(rule='Y').mean()
count = frame.resample(rule='Y').sum()
count
```

# 可视化


```python
import pandas as pd
import numpy as np

frame = np.random.randn(1000)
frame = pd.DataFrame(frame, index=pd.date_range('20200101', periods=1000))
frame['cumsum'] = frame.cumsum()
frame.plot()
frame.plot.hist(bins=50)
```

```python
import pandas as pd
import numpy as np

df = pd.DataFrame(np.random.randn(1000, 4), columns=['A', 'B', 'C', 'D'], index=pd.date_range('20200101', periods=1000))
df.cumsum()
df.plot()
```
